```python
%matplotlib inline
import pyfits
import numpy as np
import matplotlib.pyplot as plt

__author__ = 'fgh'


from astropy.io import fits


# Initial file to start all the proces,
# at the begining this script include parts that then it will be moved to other parts


# REFERENCES:
# https://github.com/shkw0k/cookbook/tree/master/


# Try to display a FIT:
# http://nbviewer.ipython.org/github/shkw0k/cookbook/blob/master/Display%20Fits.ipynb
# install packager:
#                     pip install "ipython[notebook]"

hdus = pyfits.open("/home/fgh/compartida/Astronomia/calibrated/images/20141226/zonaPolar_0_1_R_000046072.FIT")
data = hdus[0].data
print (data.shape, "min=", data.min(), "max=", data.max())
fig = plt.figure(figsize=(15, 5))
plt.subplot(141)
im = plt.imshow(data)
im.axes.set_xlabel("Default cmap")

plt.subplot(142)
im = plt.imshow(data, cmap="gray")
im.axes.set_xlabel("Gray color map")

plt.subplot(143)
im = plt.imshow(data, cmap="gray_r")
im.axes.set_xlabel("Reversed gray cmap")

plt.subplot(144)
im = plt.imshow(data, cmap="gray_r", origin='lower')
im.axes.set_xlabel("Y origin bottom left")

fig = plt.figure(figsize=(13,4))
plt.subplot(121)
h = plt.hist(data.flatten(), bins=256)

ax = plt.subplot(122)
hist, bins = np.histogram(data.flatten(), bins=256, density=False)
ax.fill(bins[:-1], hist, "r", antialiased=True)
a = ax.set_xticks(range(2000, 20001, 2000))




hdus = pyfits.open("/home/fgh/compartida/Astronomia/calibrated/images/20141229/WASP43b_R_000046395.FIT")
data = hdus[0].data
print (data.shape, "min=", data.min(), "max=", data.max())
fig = plt.figure(figsize=(15, 5))
plt.subplot(141)
im = plt.imshow(data)
im.axes.set_xlabel("Default cmap")

plt.subplot(142)
im = plt.imshow(data, cmap="gray")
im.axes.set_xlabel("Gray color map")

plt.subplot(143)
im = plt.imshow(data, cmap="gray_r")
im.axes.set_xlabel("Reversed gray cmap")

plt.subplot(144)
im = plt.imshow(data, cmap="gray_r", origin='lower')
im.axes.set_xlabel("Y origin bottom left")

fig = plt.figure(figsize=(13,4))
plt.subplot(121)
h = plt.hist(data.flatten(), bins=256)

ax = plt.subplot(122)
hist, bins = np.histogram(data.flatten(), bins=256, density=False)
ax.fill(bins[:-1], hist, "r", antialiased=True)
a = ax.set_xticks(range(2000, 20001, 2000))
```
